# Hitters Dataset

```python
df1=pd.read_csv('C:/Users/guruk/Downloads/hitters.csv')
df1
```

 Find out no of missing values in each column?
 

```python
df1.isna().sum()
```

Find out people who have batted for more than 400 matches?


```python
df1[(df1.CAtBat>400)][['Name','CAtBat']]
```

Find out the name of the player who has hit max home runs?


```python
(df1.sort_values(by=['HmRun'],ascending=False).head(1))[['Name','HmRun']]
```

Find out league wise total number of players?


```python
df1.groupby(by=['League'])[['Name']].count()
```

Find out division wise total amount of salary invested?


```python
df1.groupby(by=['Division'])[['Salary']].sum()
```

Find out five most senior players?


```python
df1.sort_values(by=['Years'],ascending=False).head(5)[['Name','Years']]
```
